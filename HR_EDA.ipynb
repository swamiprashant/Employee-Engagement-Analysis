```python
z = df['PerformanceScore'].value_counts()
plt.figure(figsize=(10, 6))

sns.lineplot(data = z, marker= 'o', color = 'purple', markersize = 10)

plt.title('PerformanceScore of employees', fontsize = 16)
plt.xlabel('PerformanceScore', fontsize=12)
plt.ylabel('values', fontsize = 12)
plt.grid()
plt.show()
```
